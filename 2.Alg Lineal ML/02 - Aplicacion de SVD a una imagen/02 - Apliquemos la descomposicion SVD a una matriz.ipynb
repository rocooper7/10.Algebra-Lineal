```python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')
imagen = Image.open(".\\imagenes\\imagen_ejemplo_frida_bredesen.jpg")
plt.imshow(imagen);
```

```python
imagen_gr = imagen.convert('LA')

imagen_mat = np.array(list(imagen_gr.getdata(band=0)), float)    #Convertir imagen a una matriz

imagen_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
```

```python
U, D, V = np.linalg.svd(imagen_mat)
```

```python
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_recons = np.matrix(U[:, :1]) * np.diag(D[:1]) * np.matrix(V[:1,:])   #Asi visualizamos la imagen con menor calidad
plt.imshow(imagen_recons, cmap='gray');
```

```python
i = 5
imagen_recons = np.matrix(U[:, :i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
plt.imshow(imagen_recons, cmap='gray')

titulo = f'valores singulares = {i}'

plt.title(titulo)
plt.show();
```
