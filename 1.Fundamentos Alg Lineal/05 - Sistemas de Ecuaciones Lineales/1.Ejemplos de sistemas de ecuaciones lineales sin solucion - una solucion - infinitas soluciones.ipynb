```python
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-6, 6)        #Sin solucion (Sobredeterminado, mas de tres ecuaciones)

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x + 1
```

```python
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```

```python
x = np.arange(-6,6)     #Con solucion  (Determinado, 2 ecuaciones)

y_2 = -1*x+3
y_3 = 2*x+1
```

```python
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```

```python
x = np.arange(-6,6)         #Soluciones infinitas  ((Indeterminado, 1 ecuaciones))

y_3 = 2*x+1
```

```python
plt.figure()

plt.plot(x, y_3)

plt.xlim(-8,8)
plt.ylim(-8,8)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```
